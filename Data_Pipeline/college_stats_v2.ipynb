```python
import pandas as pd
```

```python
def season_stats_scrape(year_start,year_end):

    # Cycle through years of interest
    for year in range(year_start,year_end+1):

        #Create empty dataframe to hold all scraped info
        full_stats_df = pd.DataFrame(columns=['#', 'Player', 'Team', 'GP', 'MPG', 'FGM', 'FGA', 'FG%', '3PM', '3PA', '3P%', 'FTM', 'FTA',
        'FT%','TOV', 'PF', 'ORB', 'DRB', 'RPG', 'APG', 'SPG', 'BPG', 'PPG'])

        # For-loop to go through players in top 2000 by ppg
        for page in range(1,21):
            url = f"https://basketball.realgm.com/ncaa/stats/{year}/Averages/Qualified/All/Season/All/points/desc/{page}/"
            
            df = pd.read_html(url)
            stats_df = df[0]

            full_stats_df = full_stats_df.append(stats_df, ignore_index=True)
            print(f"Complete page number: {page}")


        full_stats_df["Player"] = full_stats_df["Player"].replace(',','')

        # Save as CSV file with season year
        full_stats_df.to_csv(f'./data/{year-1}-{year}stats.csv',index=False)
        print(f"Completed scraping stats for the {year-1}-{year} season")
```

```python
season_stats_scrape(2007,2020)
```

```python
#Create empty dataframe to hold all scraped info
full_stats_df = pd.DataFrame(columns=['#', 'Player', 'Team', 'GP', 'MPG', 'FGM', 'FGA', 'FG%', '3PM', '3PA', '3P%', 'FTM', 'FTA', 'FT%', 'TOV', 'PF', 'ORB', 'DRB', 'RPG', 'APG', 'SPG', 'BPG', 'PPG'])


for page in range(1,21):
    url = f"https://basketball.realgm.com/ncaa/stats/2018/Averages/Qualified/All/Season/All/points/desc/{page}/"
    
    df = pd.read_html(url)
    stats_df = df[0]

    full_stats_df = full_stats_df.append(stats_df, ignore_index=True)
    print(f"Complete page number: {page}")

full_stats_df
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(full_stats_df)
```
